```python
def column_to_discrete(column_df):
    for i in range(len(column_df)):
        if column_df[i]<=33:
            column_df.loc[i] = "L"
        elif column_df[i] >=66 :
            column_df.loc[i] = "H"
        else:
            column_df.loc[i] = "M"
```

```python
column_to_discrete(df[8])
column_to_discrete(df[9])
column_to_discrete(df[21])
column_to_discrete(df[22])
```
